```python
import numpy as np
import matplotlib.pyplot as plt
from TFNet.TFNet import TFNet
from TFNet.sys_gen import SystemGenerator
from TFNet.utils import plot_results
import warnings
warnings.filterwarnings('ignore')
```

## Generating one system per class

```python
sysgen = SystemGenerator()
responses = np.zeros([5,30])
sys_true_list = []
for i in range(5):
    responses[i,:],sys_true = sysgen(i)
    sys_true_list.append(sys_true)
```

## Identification using TFNet

```python
tfnet = TFNet()
sys_pred = tfnet(responses)
```

## Plotting results

```python
plot_results(sys_true_list,sys_pred,responses)
plt.savefig("results.pdf")
```
